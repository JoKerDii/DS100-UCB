### HW4. Simulating 10,000 rolls of pair of dice

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def faired_die():
    random_num = random.random()
    if random_num < 1/6:
        return 1
    elif random_num < 2/6:
        return 2
    elif random_num < 3/6:
        return 3
    elif random_num < 4/6:
        return 4
    elif random_num < 5/6:
        return 5
    else:
        return 6
```

```python
def loaded_die():
    random_num = random.random()
    if random_num < 1/16:
        return 1
    elif random_num < 2/16:
        return 2
    elif random_num < 5/16:
        return 3
    elif random_num < 8/16:
        return 4
    elif random_num < 12/16:
        return 5
    else:
        return 6
```

```python
Y_list = []
for i in range(10000):
    X1 = faired_die()
    X2 = loaded_die()
    Y = X1*X2
    Y_list.append(Y)
    
y_mean = np.mean(Y_list)
y_var = np.var(Y_list)
print('mean: '+ str(y_mean) + '\nvariance: ' + str(y_var))
```

```python
fig = plt.figure()
plt.hist(Y_list,bins = 36)
plt.xlabel('Y')
plt.ylabel('frequency')
plt.title('Distribution of Y by Simulating 10,000 Rolls of the Pair of Dice')
fig.savefig("simulating.png")
```
